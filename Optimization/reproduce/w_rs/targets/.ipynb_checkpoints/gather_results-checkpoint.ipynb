```python
import numpy as np
import pandas as pd
%matplotlib notebook

import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import seaborn as sns; sns.set(style='whitegrid')
import pickle

from scipy.integrate import trapz
from  scipy.signal import find_peaks, peak_widths
```

```python
# unpack spectra
spec = pickle.load(open('spectra_mid.pkl', 'rb'))
print(spec[0].keys())
spec[0]['spec'].head()
```

```python
plt.plot(spec[2]['spec']['x_nm'], spec[2]['spec']['ext_norm'])
plt.plot(spec[2]['spec']['x_nm'], spec[2]['spec']['fluo_psd_norm'])
```

```python
# function to compute the overlap here
def overlap(absorb, emission):
    return np.sum(absorb * emission) / np.sqrt(np.sum(absorb * absorb) * np.sum(emission * emission))

overlap(spec[2]['spec']['ext_norm'], spec[2]['spec']['fluo_psd_norm'])
```

```python
#  function to calculate peak scores
def peak_scorer(roi, x_nm, spectra):

    # area in the roi
    roi_spectra = spectra[ x_nm<roi[1] ][ x_nm >= roi[0] ]
    roi_x_nm = x_nm[ x_nm < roi[1] ][ x_nm >= roi[0] ]
    roi_area = trapz(roi_spectra,roi_x_nm)
    tot_area = trapz(spectra, x_nm)

    return roi_area/tot_area
```

```python
# properties --> fluorescence rate in ns 
spec[100]['transtn']['fluo_rate_ns'].tolist()[-1]
```

```python
peak_scorer([400.0, 460.0], spec[1000]['spec']['x_nm'], spec[1000]['spec']['ext_norm'])
```

## make the dataframe

```python
mols = {'smiles': [], 'overlap': [], 'fluo_rate_ns': [], 
        'fluo_peak_1': [], 'fluo_peak_2': [], 
        'ext_peak_1': [], 'ext_peak_2': []}
for _, mol in enumerate(spec):
    smiles = mol['smiles'].strip()
    ovlp = overlap(mol['spec']['ext_norm'], mol['spec']['fluo_psd_norm'])
    fluo_rate = mol['transtn']['fluo_rate_ns'].tolist()[-1]
    
    fluo_peak_1 = peak_scorer([400.0, 460.0], mol['spec']['x_nm'], mol['spec']['fluo_psd_norm'])
    fluo_peak_2 = peak_scorer([670.0, 830.0], mol['spec']['x_nm'], mol['spec']['fluo_psd_norm'])
    
    ext_peak_1 = peak_scorer([400.0, 460.0], mol['spec']['x_nm'], mol['spec']['ext_norm'])
    ext_peak_2 = peak_scorer([670.0, 830.0], mol['spec']['x_nm'], mol['spec']['ext_norm'])
    
    mols['smiles'].append(smiles)
    mols['overlap'].append(ovlp)
    mols['fluo_rate_ns'].append(fluo_rate)
    
    mols['fluo_peak_1'].append(fluo_peak_1)
    mols['fluo_peak_2'].append(fluo_peak_2)
    
    mols['ext_peak_1'].append(ext_peak_1)
    mols['ext_peak_2'].append(ext_peak_2)
    
    
    
```

```python
prop = pd.DataFrame(mols)
print(prop.shape)
prop.head()
```

```python
# visualize the overlap vs. fluoresence rate
sns.scatterplot(data=prop, x='overlap', y='fluo_rate_ns')
```
